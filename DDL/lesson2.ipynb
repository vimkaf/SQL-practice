## Order By - ASC or DESC order

Syntax: SELECT * FROM table conditions ORDER BY column ASC|DESC

```sql
SELECT * FROM customers WHERE country='China' AND age >= 50 ORDER BY age ASC;
```

## COUNT, AVG, SUM

## LIKE
Like is used to search for a patterns in records/rows

```sql
SELECT * FROM customers WHERE customerName LIKE 'Tommy';
```

```sql
SELECT * FROM customers WHERE customerName LIKE '%Tommy%';
```

#### WHERE CustomerName LIKE 'a%'	Finds any values that starts with "a"
#### WHERE CustomerName LIKE '%a'	Finds any values that ends with "a"
#### WHERE CustomerName LIKE '%or%'	Finds any values that have "or" in any position
#### WHERE CustomerName LIKE '_r%'	Finds any values that have "r" in the second position
#### WHERE CustomerName LIKE 'a_%_%'	Finds any values that starts with "a" and are at least 3 characters in length
#### WHERE ContactName LIKE 'a%o'	Finds any values that starts with "a" and ends with "o"

# IN

It is used to search for an expression in a set

```sql
SELECT * FROM customers WHERE country = 'China' OR country = 'Portugal' OR country = 'Brazil';
```

```sql
SELECT * FROM customers WHERE country IN ('China','Portugal','Brazil');
```

# Between

```sql
SELECT * FROM customers WHERE age BETWEEN 20 AND 30;
```

# ALIAS - AS

#### Aliases are used to give a table, or a column in a table, a temporary name.

#### Aliases are often used to make column names more readable.

#### An alias only exists for the duration of that query.

#### An alias is created with the AS keyword.

```sql
SELECT MIN(age) FROM customers;
```

```sql
SELECT MIN(age) AS minimum_age FROM customers;
```

```sql
SELECT *, age + 1  AS next_age FROM customers LIMIT 10;
```

# GROUP BY

#### The GROUP BY statement groups rows that have the same values into summary rows, like "find the number of customers in each country".

#### The GROUP BY statement is often used with aggregate functions (COUNT(), MAX(), MIN(), SUM(), AVG()) to group the result-set by one or more columns.



```sql
SELECT country, COUNT(country) AS number FROM customers GROUP BY country;
```

```sql
SELECT age, count(age) AS number FROM customers WHERE age > 30 GROUP BY age;
```
